```
from nltk.chat.util import Chat, reflections
pairs = [
    [
        r'my name is (.*)', 
        ['hi %1',]
    ], ['(hi|hello|hey|hola|heya)', ['hi dear customer. Do you want to book an appointment, I will guide you through here? type: book now',]],
    ['(book now|book|BOOK NOW| Book|Book Now|Book now)', ['thank you for choosing us, what you would like to have haircut or massage ? for haircut type: h to try our relaxing massage services type: m',]],
    ['(At what time| what time is it|time)', ['it is 30 mins from now. any other service you want booking for?']],
    ['(h)', ['which haircut you would like? Classics? type: Capital C. Crop Top? type: CT. or try our Neckline cut, type: NC.']],
    ['(C)', ['Welcome dear customer, Our Classics wil cost you 100 bucks? should I book now? to confirm, type: yes']],
    ['(NC)', ['Welcome dear customer, Our Neckline Cut will cost you 100 bucks? should I book now? to confirm type: yes']],
    ['(CT)', ['Welcome dear customer, Our Crop Top wil will cost you 100 bucks? should I book now? to confirm type: yes']],
    ['(yes|YEs|YES|Yes)', ['Booked!, greetings from PREMIER SALOON for booking your appointment, have wounderful experience!']],
    ['(m)', ['body massage? type: BM. head massage? type: HM']],
    ['(BM|bm|bM|Bm)', ['Welcome dear customer, Our body massage is priced at INR 250? should I book now? to confirm type: yes']],
    ['(HM|hm|hM|Hm)', ['Welcome dear customer, Our head massage is priced at INR 250? should I book now? to confirm type: yes']]
]
```

```
dummy_reflections = {
    'go' : 'gone',
    'hello' : 'hey there',
}
```

```
# ok_bot.py

def ok_bot():
    print("Hi This is Premier Saloon ChatBot Assistant.Now Book appointment at your ease to avoid hassle. How may I help you? ")
    chat = Chat(pairs, reflections)

    chat.converse()
if __name__ == "__main__":
    ok_bot()
```
